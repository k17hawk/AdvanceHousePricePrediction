```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.pandas.set_option('display.max_columns', None)
df = pd.read_csv('train.csv')
df.head()
```

```python
#altering the data leakage by
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df,df['SalePrice'],test_size=0.2,random_state=100)
```

# Handling Missing values of Categorical variable

```python
features_nan=[feature for feature in df.columns if df[feature].isnull().sum()>0 and df[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),5)))
```

```python
## Replace missing value with a new label
def replace_cat_feature(df,features_nan):
    data=df.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

df=replace_cat_feature(df,features_nan)

df[features_nan].isnull().sum()
```

# Handling  numerical Feature

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[feature for feature in df.columns if df[feature].isnull().sum()>0 and df[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(df[feature].isnull().mean(),4)))
```

```python
## Replacing the numerical Missing Values using median ,because of presence of outliers

for feature in numerical_with_nan:
    median_value=df[feature].median()
    df[feature].fillna(median_value,inplace=True)
    
df[numerical_with_nan].isnull().sum()
```

# Handling  Temporal Feature

```python
##now calculating the years of house either built or modififed
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    df[feature]=df['YrSold']-df[feature]
```

```python
df.head()
```

```python
##on the basis of information obtained from EDA we perform log normal distribution to some continuous variables
continuous_feature=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in continuous_feature:
    df[feature]=np.log(df[feature])
```

# Handling Rare Categorical Feature

```python
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
```

```python
#identifying categorical variable that are present less then 1%
for feature in categorical_features:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    print(temp)
```

```python
categorical_features
```

```python
for feature in categorical_features:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'Rare_variable')
```

```python
df.head(30)
```

```python
##now replacing the categorical variable by its index number
for feature in categorical_features:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    print(labels_ordered)
```

```python
##now replacing the categorical variable  by its index number
for feature in categorical_features:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```
